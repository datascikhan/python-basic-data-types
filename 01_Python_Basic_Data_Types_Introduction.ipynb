
# Python Basic Data Types

Python me **data types** define karte hain ki ek variable me **kis type ka data** store hoga aur us par kaun se **operations** perform kiye ja sakte hain.  

Python me common **basic data types** ye hain:  

1. **int** → Integer numbers (e.g., `10`, `-5`)  
2. **float** → Floating point numbers / decimals (e.g., `3.14`, `-0.5`)  
3. **str** → Strings / text (e.g., `"Hello"`, `'Python'`)  
4. **bool** → Boolean values (`True` / `False`)  
5. **complex** → Complex numbers (e.g., `2 + 3j`)  

**Key Points:**  
- Python **dynamically typed language** hai, iska matlab hai aap variable declare karte waqt **type specify nahi karte**; Python automatically data type detect kar leta hai.  
- Data types ka use **memory allocation, type-specific operations, aur program logic** ke liye hota hai.  
- Ye **basic building blocks** hain jo aage ke advanced data structures (lists, tuples, dictionaries) ke liye foundation provide karte hain.
