# 1️⃣ Numbers (Numeric Data Types)

**Explanation:**  
Python me **numeric data types** numbers ko represent karte hain aur in par **arithmetic operations** perform kiye ja sakte hain.  
Python ke 3 main numeric types hain:  

| Type    | Example    | Description          |
|---------|------------|--------------------|
| int     | x = 10     | Integer numbers    |
| float   | y = 3.14   | Decimal numbers    |
| complex | z = 2 + 3j | Complex numbers    |

- **int** → Positive, negative integers aur zero  
- **float** → Decimal numbers, scientific notation allowed (e.g., 1.5e3 = 1500)  
- **complex** → Complex numbers jisme real aur imaginary part hota hai (`a + bj`)  

Numeric types par aap **+ - * / % // ** operations** aur built-in functions (`abs()`, `pow()`) use kar sakte hain.

---


## 1️⃣ Integer (`int`)


- Integer type whole numbers ko represent karta hai, positive, negative aur zero sab allowed hain.  



```python
# Example 1: Basic integer
x = 10
print("x =", x)
```

```python
# Example 2: Negative integer
y = -5
print("y =", y)
```

```python
# Example 3: Arithmetic operations
a = 7
b = 3
print("Sum:", a + b)
print("Difference:", a - b)
print("Product:", a * b)
print("Division:", a // b)  # Floor division
print("Modulus:", a % b)
```

```python

# Example 4: Type conversion
num = 15
print("Float:", float(num))
```

```python
# Example 5: Built-in functions
num = -20
print("Absolute:", abs(num))
print("Power:", pow(num, 2))
```

## 2️⃣ Float (float)


Float type decimal numbers ko represent karta hai. Scientific notation bhi allowed hai.


```python
# Example 1: Basic float
x = 3.14
print("x =", x)

```

```python
# Example 2: Negative float
y = -0.75
print("y =", y)


```

```python
# Example 3: Arithmetic operations
a = 2.5
b = 1.2
print("Sum:", a + b)
print("Product:", a * b)
print("Division:", a / b)

```

```python
# Example 4: Type conversion
num = 10
print("Float of integer:", float(num))

```

```python
# Example 5: Built-in functions
num = -4.5
print("Absolute value:", abs(num))
print("Power:", pow(num, 3))
```

## 3️⃣ Complex (complex)

Complex type real + imaginary numbers ko represent karta hai (a + bj).



```python
# Example 1: Basic complex number
z = 2 + 3j
print("z =", z)
```

```python
# Example 2: Access real and imaginary parts
print("Real part:", z.real)
print("Imaginary part:", z.imag)
```

```python
# Example 3: Arithmetic operations
z1 = 1 + 2j
z2 = 3 - 1j
print("Addition:", z1 + z2)
print("Multiplication:", z1 * z2)
```

```python
# Example 4: Conjugate
z = 4 + 5j
print("Conjugate:", z.conjugate())

```

```python
# Example 5: Type conversion (real to complex)
x = 5
y = complex(x)
print("Complex of x:", y)
```
