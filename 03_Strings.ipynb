## **Single-line Strings**




```python
# Example 1: Single quotes
s1 = 'Hello'
print(s1)

```

```python

# Example 2: Double quotes
s2 = "Python"
print(s2)

```

```python
# Example 3: Combining quotes
s3 = "It's Python"
print(s3)

```

```python
# Example 4: Empty string
s4 = ''
print("Empty string:", s4)

```

```python
# Example 5: String with numbers
s5 = "Python 3.10"
print(s5)
```

## **Multi-line Strings**


- Multi-line strings ke liye triple quotes ''' ''' ya """ """ ka use hota hai.
- Ye strings multiple lines me likhi ja sakti hain aur formatting preserve hoti hai.


```python

# Example 1: Using triple single quotes
s1 = '''Hello
Python
World'''
print(s1)


```

```python
# Example 2: Using triple double quotes
s2 = """This is
a multi-line
string"""
print(s2)
```

```python

# Example 3: Multi-line with indentation
s3 = """Line1
    Line2 indented
Line3"""
print(s3)

```

```python
# Example 4: Multi-line with variables
name = "Alice"
s4 = f"""Hello {name}
Welcome to Python"""
print(s4)

```

```python
# Example 5: Multi-line docstring style
s5 = """This is a docstring example
used for documentation."""
print(s5)
```

## **String Operations**

- Python me strings par operations perform kar sakte hain: concatenation, repetition, indexing, slicing.




### **1. Concatenation (+)**


- (+) operator ka use do ya zyada strings ko combine karne ke liye hota hai.



```python

# Example 1
s1 = "Hello"
s2 = "World"
print(s1 + " " + s2)

```

```python
# Example 2
a = "Python"
b = "3.10"
print(a + b)

```

```python
# Example 3
greeting = "Hi, " + "Alice!"
print(greeting)

```

```python
# Example 4
s = "Python" + " " + "Programming"
print(s)
```

```python

# Example 5
part1 = "Good"
part2 = "Morning"
message = part1 + " " + part2
print(message)
```

## **2. Repetition (*)**


- (*) operator ka use string ko multiple times repeat karne ke liye hota hai.


```python

# Example 1
s = "Hi! "
print(s * 3)
```

```python

# Example 2
line = "-" * 10
print(line)

```

```python
# Example 3
emoji = "üòä"
print(emoji * 5)

```

```python
# Example 4
word = "Python "
print(word * 2)

```

```python
# Example 5
pattern = "*-*"
print(pattern * 4)
```


## **3. Indexing (name[0])**

- Indexing ka use string ke individual character ko access karne ke liye hota hai.

- Python me indexing 0 se start hoti hai.

- Negative indexing se string end se access hota hai (-1 last character).





```python

# Example 1: Access first character
s = "Python"
print(s[0])
```

```python
# Example 2: Access last character
print(s[-1])

```

```python
# Example 3: Access third character
print(s[2])

```

```python
# Example 4: Using variable
char_index = 4
print(s[char_index])

```

```python
# Example 5: Negative indexing
print(s[-3])
```


## **4. Slicing (name[0:3])**

- Slicing ka use string ke subset of characters ko extract karne ke liye hota hai.

- Syntax: string[start:end:step]

- start ‚Üí starting index (inclusive)

- end ‚Üí ending index (exclusive)

- step ‚Üí optional, character skip karne ke liye



```python
# Example 1: Basic slice
s = "Python"
print(s[0:3])  # 'Pyt'


```

```python
# Example 2: Slice from index 2 to 5
print(s[2:5])  # 'tho'


```

```python
# Example 3: Slice with step
print(s[0:6:2])  # 'Pto'

```

```python
# Example 4: Slice till end
print(s[3:])  # 'hon'

```
