# 3️⃣ **Boolean**

Python me **Boolean data type** do values represent karta hai: `True` ya `False`.  
- Boolean values ka use **logical operations, conditions, aur control flow** me hota hai.  
- Boolean values **integers ke saath bhi kaam karte hain**, `True = 1` aur `False = 0`.  
- Boolean expressions **comparison operators** (`==, !=, >, <, >=, <=`) aur **logical operators** (`and, or, not`) se create kiye ja sakte hain.

---


```python

# Example 1: Basic Boolean values
a = True
b = False
print(a, b)

```

```python
# Example 2: Boolean from comparison
x = 10
y = 5
print(x > y)   # True
print(x == y)  # False

```

```python
# Example 3: Logical operators
a = True
b = False
print(a and b)  # False
print(a or b)   # True
print(not a)    # False

```

```python
# Example 4: Boolean as integer
print(int(True))   # 1
print(int(False))  # 0
print(True + True) # 2

```

```python
# Example 5: Boolean in condition
x = 15
if x % 2 == 0:
    print("Even")
else:
    print("Odd")
```
