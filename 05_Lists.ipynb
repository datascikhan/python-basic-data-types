# 4️⃣ **Lists**


Python me **list** ek **ordered collection** hai jo **mutable** hota hai, iska matlab hai aap **elements add, remove, aur modify** kar sakte ho.  
- Lists me **duplicate elements** allowed hain.  
- List ke elements **different data types** ke bhi ho sakte hain (integers, strings, floats, booleans, even other lists).  
- List **square brackets `[ ]`** me define ki jati hai aur elements ko **comma `,` se separate** kiya jata hai.  

**Key Points:**  
- Ordered → Index ke through elements access kiye ja sakte hain  
- Mutable → Elements ko change kar sakte hain  
- Allows duplicates → Same value multiple times ho sakti hai  
- Flexible → Different data types ek list me ho sakte hain

```python
# Example 1: Basic list
my_list = [1, 2, 3, 4, 5]
print(my_list)

```

```python
# Example 2: List with mixed data types
mixed_list = [10, "Python", 3.14, True]
print(mixed_list)
```

```python

# Example 3: List with duplicate elements
dup_list = [1, 2, 2, 3, 4, 4, 5]
print(dup_list)

```

```python
# Example 4: Accessing list elements (indexing)
fruits = ["apple", "banana", "cherry"]
print(fruits[0])   # First element
print(fruits[-1])  # Last element

```

```python
# Example 5: Modifying elements (mutable)
numbers = [10, 20, 30]
numbers[1] = 25
print(numbers)
```
