# 5️⃣ Tuples

**Explanation:**  
Python me **tuple** ek **ordered collection** hai jo **immutable** hota hai, iska matlab hai ek baar create karne ke baad **elements ko modify nahi kar sakte**.  
- Tuples me **duplicate elements** allowed hain.  
- Tuple ke elements **different data types** ke bhi ho sakte hain.  
- Tuples **parentheses `( )`** me define ki jati hain aur elements ko **comma `,` se separate** kiya jata hai.  

**Key Points:**  
- Ordered → Index ke through elements access kiye ja sakte hain  
- Immutable → Elements ko change nahi kiya ja sakta  
- Allows duplicates → Same value multiple times ho sakti hai  
- Flexible → Different data types ek tuple me ho sakte hain

---
Defined using ( )

```python
colors = ("red", "green", "blue")
print(colors[1])   # green
print(type(colors))
```

```python
# Example 1: Basic tuple
my_tuple = (1, 2, 3, 4, 5)
print(my_tuple)
```

```python
# Example 2: Tuple with mixed data types
mixed_tuple = (10, "Python", 3.14, True)
print(mixed_tuple)
```

```python
# Example 3: Tuple with duplicate elements
dup_tuple = (1, 2, 2, 3, 4, 4, 5)
print(dup_tuple)
```

```python
# Example 4: Accessing tuple elements (indexing)
fruits = ("apple", "banana", "cherry")
print(fruits[0])   # First element
print(fruits[-1])  # Last element
```

```python
# Example 5: Slicing a tuple
numbers = (10, 20, 30, 40, 50)
print(numbers[1:4])  # (20, 30, 40)
```
