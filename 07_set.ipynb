
# 6️⃣ Sets

**Explanation:**  
Python me **set** ek **unordered collection** hai jo **mutable** hota hai, lekin **duplicate elements allow nahi karta**.  
- Set ke elements **unique** hote hain.  
- Set me **indexing ya slicing** nahi hoti, kyunki elements unordered hote hain.  
- Set **curly braces `{ }`** me define ki jati hai, aur elements ko **comma `,` se separate** kiya jata hai.  
- Sets ka use **membership testing, duplicates removal, aur mathematical operations** (union, intersection, difference) ke liye hota hai.

**Key Points:**  
- Unordered → Indexing/slicing nahi hoti  
- Mutable → Elements add/remove kiye ja sakte hain  
- No duplicates → Same element sirf ek baar hota hai  
- Useful for set operations → union, intersection, difference


```python
# Example 1: Basic set
my_set = {1, 2, 3, 4, 5}
print(my_set)


```

```python
# Example 2: Set with mixed data types
mixed_set = {10, "Python", 3.14, True}
print(mixed_set)

```

```python
# Example 3: Set with duplicate elements (duplicates removed)
dup_set = {1, 2, 2, 3, 4, 4, 5}
print(dup_set)
```

```python

# Example 4: Adding and removing elements
fruits = {"apple", "banana"}
fruits.add("cherry")
fruits.remove("banana")
print(fruits)

```

```python
# Example 5: Set operations
set1 = {1, 2, 3}
set2 = {3, 4, 5}
print("Union:", set1 | set2)
print("Intersection:", set1 & set2)
print("Difference:", set1 - set2)
```

```python
numbers = {1, 2, 3, 3, 4}
numbers.add(5)
print(numbers)    # {1, 2, 3, 4, 5}
print(type(numbers))

```
