# 7️⃣ Dictionaries

**Explanation:**  
Python me **dictionary** ek **unordered collection** hai jo **key-value pairs** store karta hai.  
- Har element me **unique key** hoti hai aur **value** kisi bhi data type ki ho sakti hai.  
- Dictionaries **mutable** hoti hain, iska matlab hai aap **values update, add, ya delete** kar sakte hain.  
- Dictionary ko **curly braces `{ }`** me define kiya jata hai aur **key: value** ke format me elements store hote hain.  

**Key Points:**  
- Unordered → Indexing nahi hoti, keys ke through access hota hai  
- Mutable → Add, update, delete possible  
- Keys unique → Ek key sirf ek baar exist kar sakti hai  
- Values flexible → Numbers, strings, lists, tuples, even other dictionaries  
- Useful for → Fast lookups, mapping relationships, structured data



```python
person = {"name": "Alice", "age": 25, "city": "New York"}
print(person["name"])   # Alice
print(type(person))
```

```python
# Example 1: Basic dictionary
my_dict = {"name": "Alice", "age": 25, "city": "New York"}
print(my_dict)
```

```python

# Example 2: Accessing values using keys
print("Name:", my_dict["name"])
print("Age:", my_dict["age"])
```

```python

# Example 3: Adding and updating elements
my_dict["email"] = "alice@example.com"  # Adding
my_dict["age"] = 26                     # Updating
print(my_dict)

```

```python
# Example 4: Removing elements
del my_dict["city"]  # Remove key-value pair
print(my_dict)

```

```python
# Example 5: Dictionary with mixed value types
mixed_dict = {
    "numbers": [1, 2, 3],
    "boolean": True,
    "nested_dict": {"a": 1, "b": 2}
}
print(mixed_dict)
```
