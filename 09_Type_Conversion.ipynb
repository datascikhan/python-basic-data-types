
## 8️⃣ **Type Conversion (Casting)**

- int() 
- float()
- str()
- bool()
- list()
- tuple() 
- set()



Python me **type conversion** ka use variables ka data type **change** karne ke liye hota hai.  
- Ye **explicit conversion** hota hai, jisme hum **built-in functions** ka use karte hain:  
  `int()`, `float()`, `str()`, `bool()`, `list()`, `tuple()`, `set()`  
- Casting se hum **data ko compatible type me convert** kar sakte hain arithmetic, logical, ya data structure operations ke liye.


## **1. int() – Convert to Integer**





```python
# Example 1: Float to int
x = 3.14
print(int(x))  # 3
```

```python
# Example 2: String to int
y = "10"
print(int(y))  # 10
```

```python

# Example 3: Boolean to int
b = True
print(int(b))  # 1
```

## **3. str() – Convert to String**



```python
# Example 1: Integer to string
x = 100
print(str(x))  # '100'

```

```python
# Example 2: Float to string
y = 3.14
print(str(y))  # '3.14'

```

```python
# Example 3: Boolean to string
b = True
print(str(b))  # 'True'
```

## **4. bool() – Convert to Boolean**

```python
# Example 1: Integer to bool
x = 0
print(bool(x))  # False


```

```python
# Example 2: String to bool
s = "Python"
print(bool(s))  # True


```

```python
# Example 3: Empty string to bool
s2 = ""
print(bool(s2))  # False
```

## **5. list() – Convert to List**


```python
# Example 1: String to list
s = "Python"
print(list(s))  # ['P','y','t','h','o','n']
```

```python

# Example 2: Tuple to list
t = (1, 2, 3)
print(list(t))  # [1, 2, 3]

```

```python
# Example 3: Set to list
st = {5, 6, 7}
print(list(st))  # [5, 6, 7] (order may vary)
```

## **6. tuple() – Convert to Tuple**

```python
# Example 1: List to tuple
lst = [1, 2, 3]
print(tuple(lst))  # (1, 2, 3)
```

```python
# Example 2: String to tuple
s = "Python"
print(tuple(s))   # ('P','y','t','h','o','n')
```

```python
# Example 3: Set to tuple
st = {4, 5, 6}
print(tuple(st))  # (4,5,6) (order may vary)
```


## **7. set() – Convert to Set**

```python
# Example 1: List to set (removes duplicates)
lst = [1, 2, 2, 3]
print(set(lst))  # {1, 2, 3}


```

```python
# Example 2: String to set
s = "hello"
print(set(s))    # {'h','e','l','o'}

```

```python
# Example 3: Tuple to set
t = (5, 5, 6)
print(set(t))    # {5,6}
```
